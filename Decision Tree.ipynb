Bir karar ağacı, çok sayıda kayıt içeren bir veri kümesini, bir dizi karar kuralları uygulayarak daha küçük kümelere bölmek için kullanılan bir yapıdır. Yani basit karar verme adımları uygulanarak, büyük miktarlardaki kayıtları, çok küçük kayıt gruplarına bölerek kullanılan bir yapıdır.

**Karar Ağaçlarının Avantajları:**
* Anlaması ve yorumlaması kolaydır. Kullanılan ağaç yapılar görselleştirilebilir.
* Az oranda bir veri hazırlığına ihtiyaç duyar. Fakat unutulmamalıdır ki **bu model kayıp değerleri desteklememektedir**.
* Kullanılan ağacın maliyeti, ağacı eğitmek için kullanılan veri noktalarının sayısıyla logaritmiktir.
* Hem sayısal hem de kategorik verileri işleyebilir.
* Çok çıktılı problemleri ele alabilmektedirler.
* İstatistiksel testler kullanılarak bir modelin doğrulanması mümkündür.
* Karar ağaçları, parametrik olmayan bir yöntem olarak düşünülebilir. Yani uzay dağılımı ve sınıflandırma yapısı hakkında bir yaklaşıma sahip değilledir.

**Karar Ağaçlarının Dezavantajları:**

* Veriyi iyi bir şekilde açıklamayan aşırı karmaşık ağaçlar üretilebilir. Bu durumda ağaç dallanması takip edilemeyebilir.
* Ezbere öğrenme yaşanabilir (“over-fitting”). Bu problemin çözümü için model parametrelere kısıtlamalar ve budama gibi yöntemler kullanılabilir. Budama işlemi, az sayıda nesneyi barındıran yaprak düğümlerin karar ağacı grafiğinden atılmasını ifade etmektedir.
Karar ağacı algoritmalarında bölünmenin nasıl gerçekleşeceği ağacın doğruluğunu etkileyen faktörlerdendir. Sınıflandırma ve regresyon problemleri için bölünme kriterleri genel itibariyle farklıdır.

Karar ağaçları, bir düğümü iki veya daha fazla alt düğümde bölmeye karar vermek için birden fazla algoritma kullanır. Alt düğümlerin oluşturulması, alt düğümlerin homojenliğini arttırır. Başka bir ifadeyle, düğümün saflığının hedef değişkenlere göre arttığını söyleyebiliriz.

Algoritma seçimi, hedef değişkenin tipine dayanır. Karar ağaçlarında en sık kullanılan algoritmalar; kategorik değişkenler için Entropi, Gini, Sınıflandırma Hatası; sürekli değişkenler için ise En Küçük Karalere yöntemi şeklindedir.
**Entropi, verilerimizle ilgili belirsizliğin bir ölçüsüdür.** Sezgisel olarak, bir veri kümesinin yalnızca bir etiketi varsa (örneğin, her yolcu hayatta kaldı), daha düşük bir entropiye sahip olduğunu düşünebiliriz. Dolayısıyla verilerimizi, entropiyi en aza indirecek bir şekilde bölmemiz gerekmektedir. Bölünmeler ne kadar iyi olursa, tahminimiz de o kadar iyi olur.
> ### H = — Σ p(x) log p(x)
> Burada, p (x) belirli bir sınıfa ait grubun yüzdesini ve H ise entropiyi belirtmektedir.

## Gini Safsızlık nedir?

Gini safsızlık ölçüsü, karar ağacı algoritmalarında bir kök düğümden en uygun bölünmeye ve sonraki bölünmelere karar vermek için kullanılan yöntemlerden biridir. Bir karar ağacını bölmenin en popüler ve en kolay yoludur ve yalnızca ikili bölmeler yaptığı için yalnızca kategorik hedeflerle çalışır.

Gini Safsızlık aşağıdaki formül kullanılarak hesaplanır,

### Gini Safsızlığını düşürdüğümüzde, düğümün homojenliği yükselir. Saf bir düğümün (aynı sınıf) Gini Safsızlığı sıfırdır.

Gini kirliliğini hesaplamak için, 8 erkek ve 10 kız olmak üzere 18 öğrenciyi içeren bir veri kümesi örneği alalım ve bunları aşağıda gösterildiği gibi performansa göre ayıralım.

Yukarıdakilerin Gini Safsızlığının hesaplanması aşağıdaki gibi olacaktır:

Yukarıdaki hesaplamada, bölünmenin (kök düğüm) Ağırlıklı Gini Safsızlığını bulmak için, hem "Ortalamanın üstü" hem de "Ortalamanın altında" düğümler için 9 / 18'den başka bir şey olmayan alt düğümlerdeki öğrencilerin olasılığını kullandık. Her bir düğümdeki kız ve erkeklerin sayısı sınıftaki performanslarına bağlı olarak değişse de, her iki alt düğümde eşit öğrenci yok.

> **Aşağıda, Gini Impurity kullanarak bir karar ağacını bölme adımları verilmiştir:**
    > 1. Entropide / Bilgi kazanmada yaptığımıza benzer. Her bölünme için, her bir alt düğümün Gini Safsızlığını ayrı ayrı hesaplayın
    > 2. Her bölünmenin Gini Safsızlığını, alt düğümlerin ağırlıklı ortalama Gini Kirlilik olarak hesaplayın
    > 3. En düşük Gini Safsızlık değerine sahip bölmeyi seçin
    > 4. Homojen düğümler elde edene kadar 1-3 arası adımları tekrarlayın
* Karar ağacını geliştirmek için kök düğümü, ara düğümleri ve yaprak düğümünü bulmaya yardımcı olur.
* Sınıflandırma ağaçları için CART (sınıflandırma ve regresyon ağacı) algoritması tarafından kullanılır.
* Düğümdeki tüm durumlar tek bir hedefe düştüğünde minimuma (sıfır) ulaşır.

```python
import pandas as pd
```

```python
data = pd.read_csv("C:/Users/sahik/Documents/MAKINE_OGRENMESI/DATA/salaries.csv")
data.head()
```

```python
temp = data.drop("salary_more_then_100k", axis=1)
y = data.salary_more_then_100k
```

```python
temp.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
temp["company_n"] = le_company.fit_transform(temp.company)
temp["job_n"] = le_job.fit_transform(temp.job)
temp["degree_n"] = le_degree.fit_transform(temp.degree)
temp.head()
```

```python
X = temp.drop(["company","job","degree"], axis=1)
X.head()
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X,y) 
```

* Karar ağaçları, bir düğümü iki veya daha fazla alt düğümde bölmeye karar vermek için varsayılan olarak **"Gini Impurty"** algoritması kullanılır.

```python
model.score(X,y)
```

* Puan "1" oldu çünkü eğitim ve veri adımı için aynı veri setini kullanıyoruz. Bu nedenle çok iyi olacağını biliyorduk. Karmaşık veri setimiz olduğunda puanımız birden az olacaktır.
